# 3. Custom Bar Charts
## 3.1 Fill Pattern

```python
import matplotlib.pyplot as plt
import numpy as np

# Index Sizes

index_names = ['index 1', 'index 2', 'index 3', 'index 4', 'index 5', 'index 6']
index_sizes = [round(index_size/10**9, 2) for index_size in [1567145984, 1567145984, 391905280, 209010688, 391905280, 1567145984]]
fill_pattern = ['/', '\\', '|', '-', 'x', '']

width = 0.35  # the width of the bars



for i, index_name in enumerate(index_names):
    bar = plt.bar(i, index_sizes[i], width, label=index_name, hatch=fill_pattern[i])
    plt.bar_label(bar, padding=3)


# # Add some text for labels, title and custom x-axis tick labels, etc.
plt.ylabel('index size (GB)')
plt.ylim(0, 1.79)
plt.xlim(-0.5, 5.5)
plt.xticks([0, 1, 2, 3, 4, 5], index_names)
plt.legend()


plt.grid(axis = 'y')

plt.show()
```

## 3.2 Grouped Bar Charts

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
labels = ['Query 1', 'Query 2', 'Query 3', 'Query 4']
no_index = [30, 15, 20, 50]
index_1 = [22, 8, 15, 48]
index_2 = [10, 14, 3, 23]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

plt.bar(x - width, no_index, width, label='no index', hatch='/')
plt.bar(x, index_1, width, label='index 1', hatch='')
plt.bar(x + width, index_2, width, label='index 2', hatch='|')

# Add labels, title, and custom x-axis tick labels
plt.ylabel('execution time (s)')
plt.xticks([0, 1, 2, 3], labels)
plt.legend()

plt.grid(axis = 'y')

plt.show()
```

## 3.3 Stacked Bar Charts

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
labels = ['Query 1', 'Query 2', 'Query 3', 'Query 4']
optimization_time = [30, 15, 20, 50]
execution_time = [22, 0, 15, 48]
result_transmission = [10, 14, 3, 23]

exec_opt_time = [optimization_time[i] + execution_time[i] for i in range(len(optimization_time))] 

x = np.arange(len(labels))  # the label locations
width = 0.6  # the width of the bars


plt.bar(x, optimization_time, width, label='optimization time', hatch='/')
plt.bar(x, execution_time, width, bottom=optimization_time, label='execution time', hatch='')  # stacked on top
plt.bar(x, result_transmission, width, bottom=exec_opt_time, label='result transmission', hatch='|')

# Add labels, title, and custom x-axis tick labels
plt.ylabel('processing time (s)')
plt.xticks([0, 1, 2, 3], labels)
plt.legend()

plt.grid(axis = 'y')

plt.show()
```
